```python
mask = df2["LATITUDE"].isna() & df2["LONGITUDE"].isna()
cols = [
    "BOROUGH",
    "ZIP CODE",
    "LOCATION",
    "ON STREET NAME",
    "CROSS STREET NAME",
    "OFF STREET NAME",
]
subset = df2.loc[mask, cols]
missing_summary = subset.isna().mean()
required_cols = ["ON STREET NAME", "CROSS STREET NAME", "BOROUGH"]
geo_candidate_mask = mask.copy()
for col in required_cols:
    geo_candidate_mask &= df2[col].notna()
total_missing = int(mask.sum())
geo_count = int(geo_candidate_mask.sum())
remaining_after_geo_mask = mask & ~geo_candidate_mask
remaining_after_geo = int(remaining_after_geo_mask.sum())
zip_candidate_mask = remaining_after_geo_mask & df2["ZIP CODE"].notna()
zip_count = int(zip_candidate_mask.sum())
remaining_after_zip_mask = remaining_after_geo_mask & ~zip_candidate_mask
remaining_after_zip = int(remaining_after_zip_mask.sum())
borough_candidate_mask = remaining_after_zip_mask & df2["BOROUGH"].notna()
borough_count = int(borough_candidate_mask.sum())
after_borough_mask = remaining_after_zip_mask & ~borough_candidate_mask
unfillable_count = int(after_borough_mask.sum())
pipeline_summary = pd.DataFrame(
    {
        "count": [geo_count, zip_count, borough_count, unfillable_count],
        "fraction": [
            geo_count / total_missing if total_missing else 0.0,
            zip_count / remaining_after_geo if remaining_after_geo else 0.0,
            borough_count / remaining_after_zip if remaining_after_zip else 0.0,
            unfillable_count / total_missing if total_missing else 0.0,
        ],
    },
    index=[
        "geoclient_candidates",
        "zip_candidates",
        "borough_candidates",
        "unfillable",
    ],
)
lat_missing_only = df2["LATITUDE"].isna() & df2["LONGITUDE"].notna()
lon_missing_only = df2["LONGITUDE"].isna() & df2["LATITUDE"].notna()
mismatched_summary = pd.DataFrame(
    {
        "count": [int(lat_missing_only.sum()), int(lon_missing_only.sum())],
        "fraction": [lat_missing_only.mean(), lon_missing_only.mean()],
    },
    index=["lat_missing_only", "lon_missing_only"],
)
missing_summary, pipeline_summary, mismatched_summary
```

```python
valid_coords = df2["LATITUDE"].notna() & df2["LONGITUDE"].notna()

strict_keys = ["BOROUGH", "ON STREET NAME", "CROSS STREET NAME"]
strict_medians = (
    df2.loc[valid_coords, strict_keys + ["LATITUDE", "LONGITUDE"]]
    .dropna(subset=strict_keys)
    .groupby(strict_keys)[["LATITUDE", "LONGITUDE"]]
    .median()
    .astype(float)
â€‹)
strict_index = pd.MultiIndex.from_arrays([df2[col] for col in strict_keys], names=strict_keys)
strict_fill = strict_medians.reindex(strict_index)
strict_lat = strict_fill["LATITUDE"]
strict_lon = strict_fill["LONGITUDE"]
strict_mask = (
    df2["LATITUDE"].isna()
    & df2["LONGITUDE"].isna()
    & strict_lat.notna()
    & strict_lon.notna()
â€‹)
df2.loc[strict_mask, "LATITUDE"] = strict_lat[strict_mask].to_numpy()
df2.loc[strict_mask, "LONGITUDE"] = strict_lon[strict_mask].to_numpy()

simple_keys = ["BOROUGH", "ON STREET NAME"]
simple_medians = (
    df2.loc[valid_coords, simple_keys + ["LATITUDE", "LONGITUDE"]]
    .dropna(subset=simple_keys)
    .groupby(simple_keys)[["LATITUDE", "LONGITUDE"]]
    .median()
    .astype(float)
â€‹)
simple_index = pd.MultiIndex.from_arrays([df2[col] for col in simple_keys], names=simple_keys)
simple_fill = simple_medians.reindex(simple_index)
simple_lat = simple_fill["LATITUDE"]
simple_lon = simple_fill["LONGITUDE"]
simple_mask = (
    df2["LATITUDE"].isna()
    & df2["LONGITUDE"].isna()
    & simple_lat.notna()
    & simple_lon.notna()
â€‹)
df2.loc[simple_mask, "LATITUDE"] = simple_lat[simple_mask].to_numpy()
df2.loc[simple_mask, "LONGITUDE"] = simple_lon[simple_mask].to_numpy()

zip_lat_median = df2.groupby("ZIP CODE")["LATITUDE"].transform("median")
zip_lon_median = df2.groupby("ZIP CODE")["LONGITUDE"].transform("median")
zip_mask = df2["LATITUDE"].isna() & df2["LONGITUDE"].isna() & df2["ZIP CODE"].notna()
df2.loc[zip_mask, "LATITUDE"] = zip_lat_median[zip_mask]
df2.loc[zip_mask, "LONGITUDE"] = zip_lon_median[zip_mask]

borough_lat_median = df2.groupby("BOROUGH")["LATITUDE"].transform("median")
borough_lon_median = df2.groupby("BOROUGH")["LONGITUDE"].transform("median")
borough_mask = df2["LATITUDE"].isna() & df2["LONGITUDE"].isna() & df2["BOROUGH"].notna()
df2.loc[borough_mask, "LATITUDE"] = borough_lat_median[borough_mask]
df2.loc[borough_mask, "LONGITUDE"] = borough_lon_median[borough_mask]
```

```python
valid_coords = df2["LATITUDE"].notna() & df2["LONGITUDE"].notna()

# 1) Fill using BOROUGH + ON STREET + CROSS STREET medians
street_group = ["BOROUGH", "ON STREET NAME", "CROSS STREET NAME"]
street_medians = (
    df2.loc[valid_coords, street_group + ["LATITUDE", "LONGITUDE"]]
    .dropna(subset=street_group)
    .groupby(street_group)[["LATITUDE", "LONGITUDE"]]
    .median()
)
street_keys = pd.Series(
    list(zip(df2["BOROUGH"], df2["ON STREET NAME"], df2["CROSS STREET NAME"])),
    index=df2.index,
 )
street_lat_lookup = street_medians["LATITUDE"].to_dict()
street_lon_lookup = street_medians["LONGITUDE"].to_dict()
street_lat_fill = street_keys.map(street_lat_lookup)
street_lon_fill = street_keys.map(street_lon_lookup)
street_has_all = df2[street_group].notna().all(axis=1)
street_mask = (
    df2["LATITUDE"].isna()
    & df2["LONGITUDE"].isna()
    & street_has_all
    & street_lat_fill.notna()
    & street_lon_fill.notna()
 )
df2.loc[street_mask, "LATITUDE"] = street_lat_fill[street_mask].to_numpy()
df2.loc[street_mask, "LONGITUDE"] = street_lon_fill[street_mask].to_numpy()

# 2) Fill using BOROUGH + ON STREET medians
on_street_group = ["BOROUGH", "ON STREET NAME"]
on_street_medians = (
    df2.loc[valid_coords, on_street_group + ["LATITUDE", "LONGITUDE"]]
    .dropna(subset=on_street_group)
    .groupby(on_street_group)[["LATITUDE", "LONGITUDE"]]
    .median()
)
on_street_keys = pd.Series(
    list(zip(df2["BOROUGH"], df2["ON STREET NAME"])),
    index=df2.index,
 )
on_lat_lookup = on_street_medians["LATITUDE"].to_dict()
on_lon_lookup = on_street_medians["LONGITUDE"].to_dict()
on_lat_fill = on_street_keys.map(on_lat_lookup)
on_lon_fill = on_street_keys.map(on_lon_lookup)
on_street_mask = (
    df2["LATITUDE"].isna()
    & df2["LONGITUDE"].isna()
    & df2[on_street_group].notna().all(axis=1)
    & on_lat_fill.notna()
    & on_lon_fill.notna()
 )
df2.loc[on_street_mask, "LATITUDE"] = on_lat_fill[on_street_mask].to_numpy()
df2.loc[on_street_mask, "LONGITUDE"] = on_lon_fill[on_street_mask].to_numpy()

# 3) Fill using ZIP-level medians
zip_lat_median = df2.groupby("ZIP CODE")["LATITUDE"].transform("median")
zip_lon_median = df2.groupby("ZIP CODE")["LONGITUDE"].transform("median")
zip_mask = df2["LATITUDE"].isna() & df2["LONGITUDE"].isna() & df2["ZIP CODE"].notna()
df2.loc[zip_mask, "LATITUDE"] = zip_lat_median[zip_mask]
df2.loc[zip_mask, "LONGITUDE"] = zip_lon_median[zip_mask]

# 4) Fill remaining gaps with borough medians
borough_lat_median = df2.groupby("BOROUGH")["LATITUDE"].transform("median")
borough_lon_median = df2.groupby("BOROUGH")["LONGITUDE"].transform("median")
borough_mask = df2["LATITUDE"].isna() & df2["LONGITUDE"].isna() & df2["BOROUGH"].notna()
df2.loc[borough_mask, "LATITUDE"] = borough_lat_median[borough_mask]
df2.loc[borough_mask, "LONGITUDE"] = borough_lon_median[borough_mask]
```
